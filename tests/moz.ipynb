```python
from kwresearch.KWCollector import KWCloudBuilder

kcb = KWCloudBuilder(
    "data masking"
)
kws = kcb.get_related_keywords()
kws
```

```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def plot_word_cloud(keyphrases, type = "volume"):
    # Join keyphrases into a single string, replacing spaces within keyphrases with underscores
    if type == "volume":
        keyphrases_combined = {str(phrase).replace(" ", "_"): phrase.volume for phrase in keyphrases}
    else:
        keyphrases_combined = {str(phrase).replace(" ", "_"): phrase.difficulty for phrase in keyphrases}
    
    # Generate a keyphrase cloud
    wordcloud = WordCloud(width=800, height=400, random_state=21, max_font_size=110, background_color='white').generate_from_frequencies(keyphrases_combined)

    plt.figure(figsize=(10, 7))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis('off')
    plt.show()

plot_word_cloud(kws)
```

```python
plot_word_cloud(kws, type = "difficulty")
```
